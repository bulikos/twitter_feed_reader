### 1. Parse timeline with artificial parser

```python
# Initialize Parser
ip = ItemParser()
tweets = []

# Parse
if 'tweet-' in data['entryId']:
    tweets_ = ip.parse_tweet_result(data['content']['itemContent']['tweet_results']['result'])
    tweets.extend(tweets_)
elif 'home-conversation-' in data['entryId']:
    for item in data['content']['items']:
        tweets_ = ip.parse_tweet_result(item['item']['itemContent']['tweet_results']['result'])
        tweets.extend(tweets_)
else:
    print("UNEXPEXTED")
```
