### Load timeline

```python
# Load raw JSON file - this should be timeline JSON from chrome
filename = "../test/timelines_012026/timeline_2.json"

with open(filename, "r") as f:
    data = json.load(f)
```

### 1. Parse timeline with artificial parser

```python
# Artificial entry parser (tweet parser), we simplifiy it, because we are checking 

class PlaceHolderParser:
    def parse_tweet_result(self, item, source):
        return [{'size':len(str(item)),'source':source}]
```

```python
# Initialize Parser
tp = TimelineParser()

# Set placeholder parser for tweets
tp.item_parser = PlaceHolderParser()

# Parse
tweets, cursor = tp.parse(data)

print(f"Parsed {len(tweets)} items from timeline")
print(f"Next Cursor: {cursor}")
print(f"Additional stats: {tp.stats}")
```

### 2. Parse timeline with native parser

```python
# Initialize Parser
tp = TimelineParser()

# Parse
tweets, cursor = tp.parse(data)

print(f"Parsed {len(tweets)} items from timeline")
print(f"Next Cursor: {cursor}")
print(f"Additional stats: {tp.stats}")
```
